```python
# Cria um buffer de texto a partir da string `dados_populacao_ibge`.
dados_populacao_io = StringIO(dados_populacao_ibge)

# Lê os dados do buffer como um DataFrame, usando tabulação como separador.
novos_dados = pd.read_csv(dados_populacao_io, sep="\t", thousands=".", decimal=",")

# Remove linhas com NaN e atualiza `novos_dados` diretamente.
novos_dados.dropna(inplace=True)

# Exibe o DataFrame `novos_dados`.
novos_dados
```

```python
# Exibe um resumo conciso do DataFrame `novos_dados`, incluindo o tipo de índice e os tipos de coluna, valores não nulos e uso de memória.
novos_dados.info()
```

```python
# Remove espaços em branco da coluna "População(Censo de 2022)[2]" do DataFrame `novos_dados`.
novos_dados["População(Censo de 2022)[2]"] = novos_dados["População(Censo de 2022)[2]"].str.replace(" ", "")

# Remove espaços em branco da coluna "População(Censo de 2010)[3]" do DataFrame `novos_dados`.
novos_dados["População(Censo de 2010)[3]"] = novos_dados["População(Censo de 2010)[3]"].str.replace(" ", "")
```

```python
# Exibe um resumo conciso do DataFrame `novos_dados`, incluindo o tipo de índice e os tipos de coluna, valores não nulos e uso de memória.
novos_dados.info()

#As colunas relacionadas a População ainda estão em formato object
```

```python
# Converte o tipo de dados da coluna "População(Censo de 2022)[2]" no DataFrame `novos_dados` para inteiro de 64 bits (int64).
novos_dados["População(Censo de 2022)[2]"] = novos_dados["População(Censo de 2022)[2]"].astype("int64")

# Converte o tipo de dados da coluna "População(Censo de 2010)[3]" no DataFrame `novos_dados` para inteiro de 64 bits (int64).
novos_dados["População(Censo de 2010)[3]"] = novos_dados["População(Censo de 2010)[3]"].astype("int64")
```

```python
# Exibe um resumo conciso do DataFrame `novos_dados`, incluindo o tipo de índice e os tipos de coluna, valores não nulos e uso de memória.
novos_dados.info()

# As colunas de População agora estão no formato int64
```

```python
# Remove o símbolo de porcentagem (%) da coluna "% da pop. total"
novos_dados["% da pop. total"] = novos_dados["% da pop. total"].str.replace("%", "")

# Substitui vírgulas por pontos para corrigir o formato decimal
novos_dados["% da pop. total"] = novos_dados["% da pop. total"].str.replace(",", ".")
```

```python
# Converte a coluna "% da pop. total" para o tipo float após ajustes de formatação
novos_dados["% da pop. total"] = novos_dados["% da pop. total"].astype(float)

# Exibe um resumo das informações do DataFrame, incluindo o novo tipo de dados da coluna modificada
novos_dados.info()
```
