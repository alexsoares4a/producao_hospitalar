```python
# Importa a classe date do módulo datetime para trabalhar com datas
from datetime import date

# Define um dicionário para mapear abreviações de meses para seus respectivos números
meses = {
    "Jan": 1,
    "Fev": 2,
    "Mar": 3,
    "Abr": 4,
    "Mai": 5,
    "Jun": 6,
    "Jul": 7,
    "Ago": 8,
    "Set": 9,
    "Out": 10,
    "Nov": 11,
    "Dez": 12,
}

# Define uma função para converter uma string de ano e mês para um objeto date
def para_dia(ano_mes: str):
    # Extrai os quatro primeiros caracteres como ano
    ano: int = int(ano_mes[:4])
    # Extrai o mês como uma string a partir do quinto caractere
    mes_como_string: str = ano_mes[5:]
    # Usa o dicionário para mapear a string do mês para um número
    mes: int = meses[mes_como_string]
    # Retorna um objeto date usando o ano, o mês mapeado e define o dia como 1
    return date(ano, mes, 1)
```
