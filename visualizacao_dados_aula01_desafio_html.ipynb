```python
import pandas as pd
```

```python
# Para ler o arquivo HTML é necessário instalar o pacote lxml
# pip install lxml --upgrade
# pip install html5lib --upgrade
# pip install beautifulsoup4 --upgrade
```

```python
#url = "http://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o"
url = "https://www.ibge.gov.br/censo2010/apps/sinopse/index.php?dados=4&uf=oo"
#url = "https://www.ibge.gov.br/censo2010/apps/sinopse/index.php?uf=00&dados=1"
# url = "https://www.wikiwand.com/pt/Lista_de_unidades_federativas_do_Brasil_por_%C3%A1rea"

tabelas = pd.read_html(url, thousands=".", decimal=",")
```

```python
# Verifica o número de tabelas carregadas
print(f"Total de tabelas encontradas: {len(tabelas)}")

# Para visualizar as primerias linhas de cada tabela e identificar a desejada
for i, tabela in enumerate(tabelas):
    print(f"Tabela {i}")
    print(tabela.head())
    print("\n")
```

```python
# Obtendo os dados da tabela desejada Tabela de índice 0
dados_populacao_ibge = tabelas[0]

dados_populacao_ibge.head()
```

```python
# Removendo a coluna Região que não possui dados
dados_populacao_ibge.drop("Região", axis=1, inplace=True)

dados_populacao_ibge.head()
```

```python
# dados_populacao_ibge.columns.values[0] = "Unidade da Federação"
dados_populacao_ibge.rename(columns={'Grandes Regiões e Unidades da Federação': 'Unidade da Federação'}, inplace=True)

colunas = dados_populacao_ibge.columns[1:]

for coluna in colunas:
    if len(coluna) == 5:
        dados_populacao_ibge.rename(columns={coluna: coluna[:-1]}, inplace=True)
```

```python
dados_populacao_ibge
```

```python
dados_populacao_ibge.info()
```

```python

colunas = dados_populacao_ibge.columns[1:]

# Convertendo as colunas para numérico, transformando valores não convertíveis em NaN (Not a Number)
for coluna in colunas:
    dados_populacao_ibge[coluna] = pd.to_numeric(dados_populacao_ibge[coluna], errors='coerce').astype("Int64")

dados_populacao_ibge.info()
```

```python
dados_populacao_ibge.set_index("Unidade da Federação", inplace=True)
dados_populacao_ibge
```
